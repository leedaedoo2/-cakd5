```python
from konlpy.tag import Okt

okt = Okt()
text = "파싱은 프로그램을 compile하는 과정에서 특정 프로그래밍 언어가 제시하는 문법을 잘 지켜서 작성하였는지 compiler가 검사하는 것이다."
nounlist = okt.nouns(text)
morphlist = okt.morphs(text)
pos1 = okt.pos(text,norm=True,stem=True) # 정규화, 원형 출력
pos2 = okt.pos(text,norm=True,stem=True,join=True)


print(nounlist)
print(morphlist)
print(pos1)
print(pos2)
```

- 단어 빈도 분석 후 빈도수에서 Top 10 단어를 막대 그래프로 시각화

```python
f = open('./dataset/영화 기생충_review.txt','r')
text = f.read()

def cleanText(readData):
    readData = re.sub('[A-Za-z0-9]',' ',readData)
    readData = re.sub('[-=+♡♥;,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]',' ', readData)
    readData = re.sub('\n',' ',readData)
    return readData

newtext = cleanText(text)

pos3 = okt.pos(newtext,norm=True,stem=True)
pos3
```

```python
word_list = []

for word in pos3:
    if word[1] == 'Noun':
        word_list.append(word[0])
    elif word[1] == 'Verb':
        word_list.append(word[0])
    elif word[1] == 'Adjective':
        word_list.append(word[0])
    else:
        continue
        
print(word_list)
```

```python
from collections import Counter
count = Counter(word_list)
n_words = dict(count.most_common())
print(n_words)
```
