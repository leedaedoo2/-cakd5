```python
import pandas as pd
import numpy as np
movies = pd.read_csv('./dataset/ml-latest-small/movies.csv')
ratings = pd.read_csv('./dataset/ml-latest-small/ratings.csv')
print(movies.shape)
print(ratings.shape)
```

```python
rt = pd.pivot(ratings, index='userId',columns = 'movieId',values = 'rating')
rt.head()
```

```python
rating_movies = pd.merge(ratings,movies,on='movieId')
```

```python
rt2 = rating_movies.pivot_table('rating', index = 'userId', columns='title')
rt2 = rt2.fillna(0)
rt2
```

영화와 영화들 간 유사도 산출
- cosnine_similarity 함수는 행을 기준으로 서로 다흔 행을 비교해 유사도를 산출
- 영화를 기준으로 cosine_similarity를 적용하면 현재의 rating_matrix 가 행 기준이 되고 열 기준이 사용자가 된다

```python
ratings_matrix_T = rt2.transpose()
ratings_matrix_T
```

```python
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T,ratings_matrix_T)
item_sim
```

```python
item_sim_df = pd.DataFrame(data=item_sim,index = rt2.columns, columns=rt2.columns)

item_sim_df
```

```python
item_sim_df['Godfather, The (1972)'].sort_values(ascending=False)[:10]
```

```python
def predict_rating(ratings_arr,item_sim_arr):
    rating_pred = ratings_arr.dot(item_sim_arr)/np.array(np.abs(item_sim_arr).sum(axis=1))
    return rating_pred
```

```python
ratings_pred = predict_rating(rt2.values,item_sim_df.values)
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index = rt2.index, columns=rt2.columns)
ratings_pred_matrix
```

```python
from sklearn.metrics import mean_squared_error

def get_mse(pred,actual):
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred,actual)

print(get_mse(ratings_pred,rt2.values))
```

```python
def predict_rating_topsim(ratings_arr,item_sim_arr,n=20):
    pred = np.zeros(ratings_arr.shape)
    
    for col in range(ratings_arr.shape[1]):
        top_n_items = [np.argsort(item_sim_arr[:,col])[:-n-1:-1]]
        
        for row in range(ratings_arr.shape[0]):
            pred[row,col] = item_sim_arr[col,:][top_n_items].dot(ratings_arr[row,:][top_n_items].T)
            pred[row,col] /= np.sum(np.abs(item_sim_arr[col,:][top_n_items]))
            
    return pred
```

```python
ratings_pred = predict_rating_topsim(rt2.values,item_sim_df.values,n=20)
print(get_mse(ratings_pred,rt2.values))
```

```python
ratings_pred_matrix = pd.DataFrame(data=ratings_pred,index=rt2.index,columns = rt2.columns)
ratings_pred_matrix
```

```python
def get_unseen_movies(ratings_matrix,userId):
    user_rating = ratings_matrix.loc[userId,:]
    already_seen = user_rating[user_rating>0].index.tolist()
    
    movies_list =ratings_matrix.columns.tolist()
    unseen_list = [movie for movie in movies_list if movie not in already_seen]
    
    return unseen_list

def recomm_movie_by_userId(pred_df,userId,unseen_list,top_n=10):
    recomm_movies = pred_df.loc[userId,unseen_list].sort_values(ascending=False)[:top_n]
    return recomm_movies

unseen_list = get_unseen_movies(rt2,9)

recomm_movies = recomm_movie_by_userId(ratings_pred_matrix,9,unseen_list,top_n=10)
```
