```python
#두개의 숫자를 인자로 적용하면 4칙연산을 수행하는 사용자 함수를 작성한 후 5,3을 대입한 결과를 튜플로 출력하시오

def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def mul(x,y):
    return x*y
def div(x,y):
    return round(x/y)

a = int(input('첫번째 숫자> '))
b = int(input('두번째 숫자> '))

print(add(a,b), sub(a,b), mul(a,b),div(a,b))
```

```python
#factorial
def factorial(n):
    output =1
    for i in range(1,n+1):
        output *=i
        
    return output
    
factorial(1)
```

```python
# 재귀 : 자기 자신 호출
# 재귀함수를 이용해서 팩토리얼을 구하는 사용자 함수 작성, 1!, 5!, 10! 

def recursiveFactorial(n):
    if n < 0:
        return False
    elif n == 0:
        return 1
    else:
        return n*recursiveFactorial(n-1)
    
    
print(recursiveFactorial(0))
print(recursiveFactorial(5))
print(recursiveFactorial(10))
```

```python
#원의 면적과 둘레의 길이를 구하는 사용자 함수를 작성하여 반지름이 5와 10일때 구하시오

def circleCal(n):
    length = 2*3.14159*n
    area = 3.14159*n*n
    return length,area

n = 10

area = circleCal(n)[1]
length = circleCal(n)[0]

print(f'반지름 {5} > 면적 = {area:0.2f}, 둘레 = {length:0.2f}')
```

```python
#문자열 리스트 states를 정형화하여 아래와 같이 출력하세요
def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        value = value.strip('')
        value = value.rstrip('#')
        value = value.rstrip('!')
        value = value.rstrip('?')
        result.append(value)
    return result


states = ['     Alabama ','Georgia!','Georgia','georgia','Fl0rlda','south   carolina##','West virginia?']

#print(states)

strlist = clean_strings(states)
blankcount = 0
for a in range(len(strlist)):
    strlist[a] = list(strlist[a])
#print(strlist)
#print(len(strlist))
for i in range(len(strlist)):
    for j in range(len(strlist[i])):
        #print(len(strlist[i]))
            #print(strlist[i][j])
            #print(type(strlist[i][j]))
        if j ==0:
            strlist[i][j]=strlist[i][j].upper()
        elif strlist[i][j] == 'c':
            strlist[i][j] = strlist[i][j].upper()
        elif strlist[i][j] == '0':
            strlist[i][j] = 'o'
        elif strlist[i][j] == ' ':
            if blankcount ==2:
                continue
            else:
                strlist[i][j] = ''
                blankcount +=1
        elif strlist[i][j] =='v':
            strlist[i][j] ='V'
            
for b in range(len(strlist)):
    strlist[b] ="".join(strlist[b])

strlist
```

```python
# 
def getTextFreq(filename):
    with open(filename,'r') as f:
        text = f.read()
        fa = {}
        for c in text:
            if c in fa:
                fa[c] +=1
            else:
                fa[c] = 1
    return fa

ret = getTextFreq('data/mydata.txt')
ret = sorted(ret.items(), key=lambda x:x[1],reverse=True)
#print(ret)

for a in range(len(ret)):
    if ret[a][0] =='\n':
        continue
    print(f'[{ret[a][0]}] -> [{ret[a][1]}]회 나타남')
    
```

```python
#age 변수를10, 20, 30, 40, 50, 60이상 범주로 구분하여 age_cat 변수를 생성
def age_cat(age):
    cat = ''
    if age <=19:
        cat = '10대'
    elif age >=20 and age<=29:
        cat = '20대'
    elif age >=30 and age <=39:
        cat = '30대'
    elif age >=40 and age <=49:
        cat = '40대'
    elif age >=50 and age <=59:
        cat = '50대'
    elif age >=60:
        cat = '60대 이상'
    return cat


tt['age_cat'] = tt['age'].apply(lambda x:age_cat(x))
tt.age_cat
tt[['age','age_cat']].head()
```
