```python
df.info()
```

```python
def origincat(x):
    cat = ''
    if x == 'USA':
        cat = 'USA'
    elif x == 'EU':
        cat = 'EU'
    elif x == 'JAPAN':
        cat = 'JAPAN'
    return cat

df['origin'] = df.origin.apply(lambda x:origincat(x))
df
```

```python
# horsepower를 3개의 bin으로 나누고 bin_names['저출력','보통출력','고출력']으로 출력하세요
# np.histogram() 함수로 3개의 bin으로 나눔
```

```python
df5 = df.copy()
df5.horsepower.replace(np.nan,0,inplace = True)


arr = np.array(df5.horsepower)
histogram = np.histogram(arr,bins = 3)
print(histogram)
def horseCat(x):
    cat = ''
    if x >= histogram[1][0] and x <= histogram[1][1]:
        cat = '저출력'
    elif x>= histogram[1][1] and x <= histogram[1][2]:
        cat = '보통출력'
    elif x>= histogram[1][2] and x <= histogram[1][3]:
        cat = '고출력'
    return cat

df['horse_cat'] = df.horsepower.apply(lambda x : horseCat(x))
display(df['horse_cat'])
```
