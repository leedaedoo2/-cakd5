```python
#비행기 왕복 요금 계산기
import calendar
import numpy as np


def matrixReturn(matrix):
    newmatrix = matrix
    for a in range(len(matrix)):
        for b in range(len(matrix[a])):
            if matrix[a][b] == 0:
                newmatrix[a][b] = '*'
    return newmatrix

def matrixPrice(matrix,seednum):
    newmatrix = matrix
    np.random.seed(seednum)
    varPrice = list(np.random.randint(2000,size = (len(matrix)*len(matrix[0]))))
    #print(varPrice)
    for x in range(len(varPrice)):
        varPrice[x] = (varPrice[x]-1000)*100
        
    #print(varPrice)
    
    for a in range(len(matrix)):
        for b in range(len(matrix[a])):
            if matrix[a][b] != '*':
                newmatrix[a][b] = 500000+varPrice[(a*len(matrix))+b]
    return newmatrix
                



while 1:
    wantedYear = int(input('원하는 년도를 입력하시오...> '))
    wantedMonth = int(input('원하는 달을 입력하시오...> '))
    r = calendar.monthcalendar(wantedYear,wantedMonth)
    #print(r)
    monthdays = max(max(r))
    #print(monthdays)
    #print(len(r[0]))
    #print(len(r))
    rr = matrixReturn(r)
    temprr = rr
    #print(rr)
    
    rrr = matrixPrice(rr,wantedYear*100+wantedMonth)
    #print(rrr)
    
    newr = calendar.monthcalendar(wantedYear,wantedMonth)
    newrr = matrixReturn(newr)
    
    print('*'*63)
    print(' '*23+f'{wantedYear}년 {wantedMonth}월 요금표')
    print('*'*63)
    
    for a in range(len(r)):
        print('')
        for b in range(len(r[a])):
            if len(str(newrr[a][b])) < 2:
                print((' '*4)+str(newrr[a][b])+(' '*4),end = '')
            else:
                print((' '*3)+str(newrr[a][b])+(' '*4),end = '')
        print('')
        print('-'*63)
        for b in range(len(r[a])):
            if len(str(rrr[a][b])) < 2:
                print((' '*4)+str(rrr[a][b])+(' '*4),end = '')
            else:
                print(' '+str(rrr[a][b])+('원'),end = '')
        print('')
    print('')
    monthRange = []
    monthfee = []
    
    for day in range(1,monthdays+1):
        monthRange.append(day)
        
    #print(monthRange)
    
    for c in range(len(rrr)):
        for d in range(len(rrr[c])):
            if rrr[c][d] != '*':
                monthfee.append(rrr[c][d])
    #print(monthfee)
    
    feeMatrix = []
    feeDict = {}
    t = 1
    for x in range(monthdays):
        dayline = []
        for y in range(monthdays):
            if y - x < 3:
                dayline.append('*')
            else:
                dayline.append(monthfee[x] + monthfee[y])
                feeDict[t] = [monthfee[x]+monthfee[y],x+1,y+1,y-x]
                t +=1
        feeMatrix.append(dayline)        
    
    #print(feeMatrix)
    #print(feeDict)
    #print(len(feeDict))
    
    findBest5 = {}    
    minkey = 1
    for i in range(1,6):
        for j in range(1,len(feeDict)+1):
            if feeDict[minkey][0] > feeDict[j][0]:
                 minkey = j
        #print(feeDict[minkey])
        findBest5[i] = feeDict[minkey]
        
        for k in range(minkey,len(feeDict)):
            feeDict[k] = feeDict[k+1]
        del feeDict[len(feeDict)]
    
    print('')
    #print(findBest5)
    print('선택한 달의 최소 가격(최소 3일 이상의 기간이 선택됩니다)')
    
    print(f'Best1 : {findBest5[1][1]}일 ~ {findBest5[1][2]}일({findBest5[1][3]}일간), {findBest5[1][0]}원')
    print(f'Best2 : {findBest5[2][1]}일 ~ {findBest5[2][2]}일({findBest5[2][3]}일간), {findBest5[2][0]}원')
    print(f'Best3 : {findBest5[3][1]}일 ~ {findBest5[3][2]}일({findBest5[3][3]}일간), {findBest5[3][0]}원')
    print(f'Best4 : {findBest5[4][1]}일 ~ {findBest5[4][2]}일({findBest5[4][3]}일간), {findBest5[4][0]}원')
    print(f'Best5 : {findBest5[5][1]}일 ~ {findBest5[5][2]}일({findBest5[5][3]}일간), {findBest5[5][0]}원')
    print('')
    myChoice = []
    
    while 1:
        choice1 = input('원하는 일정이 있나요? [y/n]> ')
        if choice1 == 'y':
            while 1:
                choice2 = input('원하는 일정을 선택해주세요(1,2,3,4,5)> ')
                if int(choice2) in range(1,6):
                    myChoice = findBest5[int(choice2)]
                    print(f'{int(choice2)}번이 선택되었습니다.')
                    break
                else:
                    print('잘못 입력했습니다. 다시 입력해주세요')
                    continue
            break
        elif choice1 != 'y' and choice1 != 'n':
            print('잘못 입력했습니다. 다시 입력해주세요')
            continue
        else:
            while 1:
                print('')
                wantDays = input('원하는 여행기간을 입력해주세요(최소 3일 이상)> ')
                if int(wantDays) < 3:
                    print('기간이 짧습니다. 다시 입력해주세요')
                    continue
                elif int(wantDays) >= 3:
                    wantDays = int(wantDays)
                    break
                else:
                    print('잘못 입력했습니다. 다시 입력해주세요')
                    continue
            t = 1
            feeDict = {}
            for x in range(monthdays):
                for y in range(monthdays):
                    if y-x == wantDays:
                        feeDict[t] = [monthfee[x]+monthfee[y],x+1,y+1,y-x]
                        t +=1
            #print(feeDict)
            findBest5 = {}    
            minkey = 1
            for i in range(1,6):
                for j in range(1,len(feeDict)+1):
                    if feeDict[minkey][0] > feeDict[j][0]:
                         minkey = j
                        #print(feeDict[minkey])
                findBest5[i] = feeDict[minkey]
        
                for k in range(minkey,len(feeDict)):
                    feeDict[k] = feeDict[k+1]
                del feeDict[len(feeDict)]
            print('')    
            print(f'Best1 : {findBest5[1][1]}일 ~ {findBest5[1][2]}일({findBest5[1][3]}일간), {findBest5[1][0]}원')
            print(f'Best2 : {findBest5[2][1]}일 ~ {findBest5[2][2]}일({findBest5[2][3]}일간), {findBest5[2][0]}원')
            print(f'Best3 : {findBest5[3][1]}일 ~ {findBest5[3][2]}일({findBest5[3][3]}일간), {findBest5[3][0]}원')
            print(f'Best4 : {findBest5[4][1]}일 ~ {findBest5[4][2]}일({findBest5[4][3]}일간), {findBest5[4][0]}원')
            print(f'Best5 : {findBest5[5][1]}일 ~ {findBest5[5][2]}일({findBest5[5][3]}일간), {findBest5[5][0]}원')
            print('')
            while 1:
                choice3 = input('원하는 일정이 있나요? [y/n]> ')
                if choice3 == 'y':
                    choice4 = input('원하는 일정을 선택해주세요(1,2,3,4,5)> ')
                    if int(choice4) in range(1,6):
                        myChoice = findBest5[int(choice4)]
                        print(f'{int(choice4)}번이 선택되었습니다.')
                        break
                    else:
                        print('잘못 입력했습니다. 다시 입력해주세요')
                        continue
                elif choice1 != 'y' and choice1 != 'n':
                    print('잘못 입력했습니다. 다시 입력해주세요')
                    continue          
                
                break
        
        
        break
    print('')
    #print(myChoice)
    if myChoice == []:
        print('원하는 일정이 없어 아쉽네요. 나중에 다시 시도하세요')
    else:
        print(f'당신의 일정: {wantedYear}년 {wantedMonth}월 {myChoice[1]}일 ~ {wantedYear}년 {wantedMonth}월 {myChoice[2]}일')
        print(f'요금 : {myChoice[0]}원')
        print('즐거운 여행 되세요!')
    
    
    break
```
