```python
# 아래 코드를 보완하여 발생 가능한 모든 예외(정수,인덱스, 미리 파악하지 못한 예외까지)를 처리한 후 마지막에 모든 예외처리 완료를 출력하세요

list_num = [52,273,32,72,100]
while 1:
    try:
        num_input = int(input('정수 입력> '))

        print('{}번째 요소: {}'.format(num_input,list_num[num_input-1]))
        print((list_num[num_input-1])/0)
    except ValueError:
        print('정수값을 입력하세요')
    except IndexError:
        print('인덱스 오류, 범위내 값을 입력하세요')
    except ZeroDivisionError:
        print('0으로 나눠버렸어요')
    except Exception as e:
        print(e)
        print('새로운 오류: ',type(e))
    finally:
        print('모든 예외처리 완료')
```

```python
#4칙 연산 기능을 포함한 Cal4클래스(생성자 이용)을 작성하고 이 클래스를 이용하여 Cal 계산기 객체를 만든 후 적절한 수를 적용하여 사칙연산을 수행하시오

class Cal4:
    def add(self,num1,num2):
        return num1+num2
    def diff(self,num1,num2):
        return num1-num2
    def mul(self,num1,num2):
        return num1*num2
    def div(self,num1,num2):
        try:
            ret = num1/num2
            return ret
        except Exception as e:
            return e
            
obj = Cal4()
while 1:
    try:
        x = int(input("숫자 1: "))
        y = int(input("숫자 2: "))
        break
    except:
        print("올바른 정수를 입력하세요")
        continue





print(obj.add(x,y))
print(obj.diff(x,y))
print(obj.mul(x,y))
print(obj.div(x,y))
```

```python
import numpy as np


class Bank:
    bankmoney = 0
    def viewmoney(self):
        return self.bankmoney
    def deposit(self,money):
        self.bankmoney += money
        
    def withdraw(self,money):
        if self.bankmoney < money:
            print('인출 금액이 부족합니다')
            return 0
        else:
            self.bankmoney -= money
            return money
    def interest(self):
        a = np.random.randint(6)
        a = 1+0.01*a
        self.bankmoney *= a
        self.bankmoney = round(self.bankmoney)

class Coin:
    def cointoss(self):
        x = np.random.randint(2)
        return x
    def coinresult(self,x,money):
        rest = 0
        if x == 0:
            print('앞면이 나왔습니다')
            print('당첨금 두배')
            rest +=  2*money
        elif x == 1:
            print('꽝입니다')
            rest +=  0
        return rest
    
    
class Man(Bank,Coin):
    def __init__(self):
        self.budget = 50000
    

person = Man()
print('인생 게임에 온 것을 환영합니다?')
while 1:
    if person.budget == 0 and person.bankmoney == 0:
        print('HAHAHA 당신은 이제 끝장입니다!')
        break
    
    
    choice1 = input('어디로 갈까요? 1.은행 2.도박장 3.금액 확인> ')
    if choice1 not in ['1','2','3']:
        print('다시 누르세요')
        
        continue
    elif choice1 == '1':
        while 1:
            choice2 = input('무엇을 할까요? 1.저금액 확인 2.저금 3.인출 아무거나 눌러서 뒤로 갑니다> ')
            if choice2 == '1':
                print('현재 저금액 : ',person.bankmoney)
                continue
            elif choice2 == '2':
                while 1:
                    try:
                        depositmoney = input('저금할 금액을 적어주세요 (0을 눌러 취소)> ')
                        
                        if int(depositmoney) > person.budget:
                            print('금액이 부족합니다')
                        elif int(depositmoney) == 0:
                            print('저금을 취소합니다')
                            break
                        else:
                            print(f'{depositmoney}원 저금합니다')
                            person.budget -=int(depositmoney)
                            person.deposit(int(depositmoney))
                            break
                    except:
                        print('다시 적어주세요')
                        continue
                    
            elif choice2 == '3':
                while 1:
                    try:
                        withdrawmoney = input('인출할 금액을 적어주세요 (0을 눌러 취소)> ')
                        
                        if int(withdrawmoney) > person.bankmoney:
                            print('금액이 부족합니다')
                        elif int(withdrawmoney) == 0:
                            print('인출을 취소합니다')
                            break
                        else:
                            print(f'{withdrawmoney}원 인출합니다')
                            person.budget += person.withdraw(int(withdrawmoney))
                            break
                    except:
                        print('다시 적어주세요')
                        continue
            else:
                break
    elif choice1 == '3':
        print(f'현재 소지금 :{person.budget}원')
        continue
    elif choice1 == '2':
        while 1:
            print('코인 토스 도박장에 오신 것을 환영합니다!!!!!')
            choice3 = input('한번 던지실까요??> [y/n]')
            if choice3 not in ['y','n']:
                print('다시 골라주세요!')
                continue
            elif choice3 == 'n':
                print('그럼 안녕히가세요!')
                break
            else:
                while 1:
                    try:
                        bet = input('그럼 얼마를 걸어보실까요?? (0을 눌러 뒤로가기)> ')
                        
                        if int(bet) > person.budget:
                            print('이럴수가! 돈이 부족하군요!')
                            continue
                        elif int(bet) == 0:
                            print('그럼 충분히 생각해주세요!')
                            break
                        else:
                            print('그가 던졌습니다!')
                            person.budget -= int(bet)
                            coin = person.cointoss()
                            result = person.coinresult(coin,int(bet))
                            person.budget += result
                            print(f'{int(bet)}원 내고 {result}원 벌었습니다!')
                            print('그러는 동안 하루가 지났네요...(대충 이자 들어왔다는 소리)')
                            person.interest()
                            break
                    except:
                        print('잘못 누르셨네요!')
                        continue
```
