```python
#100명에 대하여 아래와 같이 출력하세요
# 이름 : 라바
# 몸무게 : 80
#키 : 146
#BMI : 35.56789    rount(int(weight)/((int(height)/100)**2),2)
#결과 : 과체중


def checkBMI(weight,height):
    status =''
    bmi = round(int(weight)/((int(height)/100)**2),2)
    if bmi < 18.5:
        status = '저체중'
    elif bmi >= 18.5 and bmi <=23:
        status = '정상'
    elif bmi >23 and bmi<=25:
        status = '과체중'
    elif bmi >25:
        status = '비만'
    return bmi,status
        
        
        
import random
hanguls = list('가나다라마바사아자차카타파하')
with open('info.txt','w') as file:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        height = random.randrange(140,200)
        weight = random.randrange(60,90)
        lookbmi = checkBMI(weight,height)
        file.write(f'이름 : {name}\n몸무게 : {weight}kg\n키 : {height}\nBMI : {lookbmi[0]}\n상태 : {lookbmi[1]}\n\n')
        
with open('info.txt','r') as file:
    data = file.read()
    print(data)
```

```python
# 주어진 데이터에서 1880년에서 2015년까지 연도별 출생아 수를 출력

for year in range(1880,2020):
    yobyear = 'yob'+str(year)
    #print(yobyear)

    with open(f'data/names/{yobyear}.txt','r') as f:
        newlist = []
        total = 0
        data = f.read()
        newdata = data.split()
        for a in newdata:
            newa = a.split(',')
            newlist.append(newa)
    
        #print(newlist)
        for b in newlist:
            total += int(b[2])
        
        print(year, total)
```

```python
print('year male   female')

for year in range(1880,2020):
    yobyear = 'yob'+str(year)
    #print(yobyear)

    with open(f'data/names/{yobyear}.txt','r') as f:
        newlist = []
        Mtotal = 0
        Ftotal = 0
        data = f.read()
        newdata = data.split()
        for a in newdata:
            newa = a.split(',')
            newlist.append(newa)
    
        #print(newlist)
        for b in newlist:
            if b[1] == 'M':
                Mtotal += int(b[2])
            elif b[1] == 'F':
                Ftotal += int(b[2])
        
        print(year, Mtotal, Ftotal)
```

```python
while 1:

    year = int(input('원하는 연도를 입력하세요(1880~2019)> '))
    print('')
    if year not in range(1880,2020):
        print('잘못 입력했습니다. 다시 입력하세요')
        print('')
        continue
    
    
    yobyear = 'yob'+str(year)
    #print(yobyear)

    with open(f'data/names/{yobyear}.txt','r') as f:
        newlist = []
        Mnames = []
        Fnames = []
        data = f.read()
        newdata = data.split()
        for a in newdata:
            newa = a.split(',')
            newlist.append(newa)

        for b in newlist:
            if b[1] == 'M':
                Mnames.append(b)
            elif b[1] =='F':
                Fnames.append(b)
        Mnames = sorted(Mnames,key=lambda x:int(x[2]),reverse = True)
        Fnames = sorted(Fnames,key=lambda x:int(x[2]),reverse = True)
        print(f'{year}년도 인기있는 이름')
        print('')
        for c in range(10):
            print(f'TOP_{c+1} 남자아기이름: ({Mnames[c][0]},{Mnames[c][2]})')
        print('')
        for d in range(10):
            print(f'TOP_{d+1} 여자아기이름: ({Fnames[d][0]},{Fnames[d][2]})')
        print('')
```
