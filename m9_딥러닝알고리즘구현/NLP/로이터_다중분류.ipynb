```python
from tensorflow.keras.datasets import reuters

(train_data, train_labels) , (test_data, test_labels) = reuters.load_data(num_words=10000)
```

```python
print(len(train_data), len(test_data))
```

```python
print(train_data[0])
```

```python
train_labels[:10]
```

```python
# keras를 이용한 방법
from tensorflow.keras.utils import to_categorical

one_hot_train_labels = to_categorical(train_labels)

one_hot_test_labels = to_categorical(test_labels)
```

```python
print(one_hot_train_labels.shape, one_hot_test_labels.shape)
```

```python
import numpy as np
def vectorize_sequence(sequences,dimension=10000):
  results = np.zeros((len(sequences),dimension))
  for i , sequence in enumerate(sequences):
    results[i,sequence] = 1.
  return results

X_train = vectorize_sequence(train_data)

X_test = vectorize_sequence(test_data)

```

```python
print(X_train.shape,X_test.shape)
```

- 모델 compile
- 훈련 검증 : 1000개 샘플을 검증 세트로 사용
- 학습 : epochs = 20, batch_size=512
- 훈련과정 시각화
- 적합한 epoch 확인

```python
# 모델 구성

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Dense, LSTM

model = Sequential()
model.add(Embedding(46, 32))
model.add(LSTM(32))
model.add(Dense(1,activation = 'sigmoid'))

model.compile(optimizer = 'rmsprop',
              loss = 'categorical_crossentropy',
              metrics = ['acc'])
```

```python
model.summary()
```

```python
history = model.fit(X_train,one_hot_train_labels, epochs = 20, batch_size = 512, validation_batch_size=1000)
```
