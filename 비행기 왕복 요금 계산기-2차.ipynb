```python
#비행기 왕복 요금 계산기
import calendar
import numpy as np

#Flight 클래스 : 항공사별 요금표 설정
class Flight():
    def __init__(self,name):
        self.name = name
    def matrixPrice(self,year,month):
        a = 0
        if self.name == 'A':
            a = 1
        elif self.name == 'B':
            a = 2
        elif self.name == 'C':
            a = 3
        matrix = calendar.monthcalendar(year,month)
        np.random.seed(year*1000+a*100+month)
        varPrice = list(np.random.randint(20000,size = (len(matrix)*len(matrix[0]))))
        for x in range(len(varPrice)):
            varPrice[x] = (varPrice[x]-10000)*10
        for a in range(len(matrix)):
            for b in range(len(matrix[a])):
                if matrix[a][b] == 0:
                    matrix[a][b] ='*'
                else:
                    matrix[a][b] = 500000+varPrice[(a*len(matrix))+b]
        return matrix

# 설정한 년월의 달력을 가공
def matrixReturn(year,month):
    matrix = calendar.monthcalendar(year,month)
    monthdays = max(max(matrix))
    for a in range(len(matrix)):
        for b in range(len(matrix[a])):
            if matrix[a][b] == 0:
                matrix[a][b] = '*'
    return matrix, monthdays

# 달력 출력
def printCalendar(matrix,a):
    print(' '*3,end = '')
    for b in range(len(matrix[a])):
        if len(str(matrix[a][b])) < 2:
            print((' '*4)+str(matrix[a][b])+(' '*4),end = '')
        else:
            print((' '*3)+str(matrix[a][b])+(' '*4),end = '')

# 항공사별 요금표 출력
def printPrice(matrix,a,name):
    print(' '+name+' ',end = '')
    for b in range(len(matrix[a])):
        if len(str(matrix[a][b])) < 2:
            print((' '*4)+str(matrix[a][b])+(' '*4),end = '')
        else:
            print(' '+str(matrix[a][b])+('원'),end = '')

# 코드별 항공사 이름 설정
def flightDetect(finput):
    if finput == 0:
        return 'A'
    elif finput == 1:
        return 'B'
    elif finput == 2:
        return 'C'
    
flightA = Flight('A')
flightB = Flight('B')
flightC = Flight('C')
while 1:
    try:
# 년월 설정
        wantedYear = int(input('원하는 년도를 입력하시오...> '))
        wantedMonth = int(input('원하는 달을 입력하시오...> '))
        r, monthdays= matrixReturn(wantedYear,wantedMonth)
# 항공사별 가격 설정
        rrr = flightA.matrixPrice(wantedYear,wantedMonth)
        rrrb = flightB.matrixPrice(wantedYear,wantedMonth)
        rrrc = flightC.matrixPrice(wantedYear,wantedMonth)
# 달력 및 요금표 출력
        print('*'*66)
        print(' '*26+f'{wantedYear}년 {wantedMonth}월 요금표')
        print('*'*66)
        for a in range(len(r)):
            print('')
            printCalendar(r,a)
            print('')
            print('-'*66)
            printPrice(rrr,a,flightA.name)
            print('')
            printPrice(rrrb,a,flightB.name)
            print('')
            printPrice(rrrc,a,flightC.name)
            print('')
        print('')
# 일자별 요금 딕셔너리 설정 [일자 : A항공사 요금, B항공사 요금, C항공사 요금]       
        monthfee = {}
        index = 1
        for c in range(len(rrr)):
            for d in range(len(rrr[c])):
                if rrr[c][d] != '*' and rrrb[c][d] != '*' and rrrc[c][d] != '*':
                    monthfee[index] = [rrr[c][d],rrrb[c][d],rrrc[c][d]]
                    index +=1
# 가격표 = [총 가격, 출발일, 도착일, 출발항공사, 도착항공사, 여행기간] 항공사: 0=A,1=B,2=C
        feeDict = {}
        t = 1
        for x in range(1,monthdays+1):
            dayline = []
            for y in range(1,monthdays+1):
                if y - x >= 3:
                    for a in range(3):
                        for b in range(3):
                            feeDict[t] = [monthfee[x][a]+monthfee[y][b],x,y,a,b,y-x]
                            feeDict[t][3] = flightDetect(feeDict[t][3])
                            feeDict[t][4] = flightDetect(feeDict[t][4])
                            t +=1       
        feeSorted = sorted(feeDict,key = lambda x:feeDict[x][0])
        print('')
        print('선택한 달의 최소 가격(최소 3일 이상의 기간이 선택됩니다)')
        print('')
        for i in range(10):
            try:
                print(f'Best{i+1} : {feeDict[feeSorted[i]][1]}일,{feeDict[feeSorted[i]][3]}항공사 ~ {feeDict[feeSorted[i]][2]}일,{feeDict[feeSorted[i]][4]}항공사({feeDict[feeSorted[i]][5]}일간), {feeDict[feeSorted[i]][0]}원')
            except IndexError:
                break
        print('')
        myChoice = []
        while 1:
            choice1 = input('원하는 일정이 있나요? [y/n]> ')
            if choice1 == 'y':
                while 1:
                    try:
                        choice2 = input('원하는 일정을 선택해주세요(1~10)> ')
                        if int(choice2) in range(1,11):
                            myChoice = feeDict[feeSorted[int(choice2)-1]]
                            print(f'{int(choice2)}번이 선택되었습니다.')
                            break
                        else:
                            print('잘못 입력했습니다. 다시 입력해주세요')
                            continue
                    except:
                        print('유효하지 않은 입력입니다 다시 입력해주세요')
                break
            elif choice1 != 'y' and choice1 != 'n':
                print('잘못 입력했습니다. 다시 입력해주세요')
                continue
            else:
                while 1:
                    try:
                        print('')
                        wantDays = input('원하는 여행기간이 있다면 입력해주세요(최소 3일 이상)> ')
                        if int(wantDays) < 3:
                            print('기간이 짧습니다. 다시 입력해주세요')
                            continue
                        elif int(wantDays) >= 3:
                            wantDays = int(wantDays)
                            break
                    except:
                        print('잘못 입력했습니다. 다시 입력해주세요')
                        continue
                t = 1
                feeDict2 = {}
                for x in range(1,monthdays+1):
                    for y in range(1,monthdays+1):
                        if y-x == wantDays:
                            for a in range(3):
                                for b in range(3):
                                    feeDict2[t] = [monthfee[x][a]+monthfee[y][b],x,y,a,b,y-x]
                                    feeDict2[t][3] = flightDetect(feeDict2[t][3])
                                    feeDict2[t][4] = flightDetect(feeDict2[t][4])
                                    t +=1
                if feeDict2 == {}:
                    print('희망 기간의 일정이 없네요. 다시 시도하세요')
                    continue                
                feeSorted2 = sorted(feeDict2,key = lambda x:feeDict2[x][0])
                print('')
                for i in range(10):
                    try:
                        print(f'Best{i+1} : {feeDict2[feeSorted2[i]][1]}일,{feeDict2[feeSorted2[i]][3]}항공사 ~ {feeDict2[feeSorted2[i]][2]}일,{feeDict2[feeSorted2[i]][4]}항공사({feeDict2[feeSorted2[i]][5]}일간), {feeDict2[feeSorted2[i]][0]}원')
                    except IndexError:
                        break
                print('')
                while 1:
                    choice3 = input('원하는 일정이 있나요? [y/n]> ')
                    if choice3 == 'y':
                        while 1:
                            try:
                                choice4 = input('원하는 일정을 선택해주세요(1~10)> ')
                                if int(choice4) in range(1,11):
                                    myChoice = feeDict2[feeSorted2[int(choice4)-1]]
                                    print(f'{int(choice4)}번이 선택되었습니다.')
                                    break
                                else:
                                    print('범위 밖의 숫자입니다. 다시 입력해주세요')
                                    continue
                            except:
                                print('유효하지 않은 입력입니다. 다시 입력해주세요')
                                continue
                        break
                    elif choice3 != 'y' and choice3 != 'n':
                        print('잘못 입력했습니다. 다시 입력해주세요')
                        continue          
                    else:
                        feeDict3 = {}
                        index = 1
                        while 1:
                            try:
                                startday = int(input('출발일자를 선택해주세요> '))
                                endday = int(input('도착일자를 선택해주세요> '))
                                if startday not in range(1,monthdays+1) or endday not in range(1,monthdays+1):
                                    print('유효범위 밖의 날짜입니다. 다시 입력해주세요')
                                    continue
                                elif endday - startday < 3:
                                    print('3일 이상의 기간이어야 하고, 도착일은 출발일보다 앞설수 없습니다.')
                                    continue
                                else:
                                    print('')
                                    print(f'당신의 선택일자 : {startday}일 ~ {endday}일')
                                    print('')
                                    for j in range(1,len(feeDict)+1):
                                        if feeDict[j][1] == startday and feeDict[j][2] == endday:
                                            feeDict3[index] = feeDict[j]
                                            index += 1
                                    feeSorted3 = sorted(feeDict3,key = lambda x:feeDict3[x][0])
                                    for k in range(len(feeDict3)):
                                        print(f'Best{k+1} : {feeDict3[feeSorted3[k]][3]}항공사 ~ {feeDict3[feeSorted3[k]][4]}항공사, {feeDict3[feeSorted3[k]][0]}원')
                                    print('')
                                    while 1:
                                        try:
                                            choice5 = int(input('원하는 일정을 설정해주세요> '))
                                            if choice5 not in range(1,len(feeDict3)+1):
                                                print('범위 밖의 숫자입니다. 다시 입력해주세요> ')
                                                continue
                                            else:
                                                myChoice = feeDict3[feeSorted3[choice5-1]]
                                                print(f'{choice5}번이 선택되었습니다.')
                                                break
                                        except:
                                            print('유효하지 않은 값입니다. 다시 입력해주세요')
                                            continue
                                    break
                            except:
                                print('유효하지 않은 값입니다. 다시 입력해주세요')
                                continue
                                
                            
                    break

            break
        print('')
        if myChoice == []:
            print('원하는 일정이 없어 아쉽네요. 나중에 다시 시도하세요')
        else:
            print(f'당신의 일정 : {wantedYear}년 {wantedMonth}월 {myChoice[1]}일 {myChoice[3]}항공사 ~ {wantedYear}년 {wantedMonth}월 {myChoice[2]}일 {myChoice[4]}항공사')
            print(f'총 여행기간 : {myChoice[5]}일')
            print(f'요금 : {myChoice[0]}원')
            print('즐거운 여행 되세요!')

    except ValueError:
        print('잘못 입력했습니다. 다시 입력해주세요')
        break
    break
```
