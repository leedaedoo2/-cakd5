```python
import pandas as pd
import numpy as np
t3 = pd.read_csv('./visualization/dataset/titanic3.csv')
t3
```

```python
# 호칭 데이터 분석
import re

p = re.compile('[A-Z][a-z]+[.]')


namelist = list(t3['name'])
callname = []

for i in range(len(namelist)):
    
    m = p.findall(namelist[i])
    callname.append(m[0])


print(len(callname))

t3['callname'] = callname
t3
```

```python
# 성씨 데이터 분석
pp = re.compile('[A-Z].+[,]')
ppp = re.compile('[A-Z].+[a-z]')
namelist = list(t3['name'])
surname = []

for i in range(len(namelist)):
    mm = pp.findall(namelist[i])
    mmm = ppp.findall(mm[0])
    surname.append(mmm[0])
    

print(len(surname))
    
t3['surname'] = surname
t3
    
```

```python
def age_cat(x):
    cat = ''
    if x<10:
        cat = 'child'
    elif x<20:
        cat = 'teen'
    elif x<30:
        cat = '20s'
    elif x<40:
        cat = '30s'
    elif x<50:
        cat = '40s'
    elif x<60:
        cat = '50s'
    else:
        cat = 'eldery'
    return cat

t3['age_cat'] = t3.age.apply(lambda x:age_cat(x))
t3.head()
```

```python
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
le = LabelEncoder()
t3['age_cat_label'] = le.fit_transform(t3['age_cat'])
t3
```

```python
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')
le = LabelEncoder()
t3.boat.fillna('0',inplace=True)
t3['boat_label'] = le.fit_transform(t3['boat'])
t3['gender_label'] = le.fit_transform(t3.sex)
t3['home.dest'].fillna('0',inplace=True)
t3['home_dest'] = le.fit_transform(t3['home.dest'])
t3['call_name'] = le.fit_transform(t3.callname)
t3
```

```python
df1 = pd.read_pickle('./visualization/dataset/tdf.pkl')
df1.age_cat = t3.age
#df1['boat'] = t3.boat_label
df1['pclass'] = t3.pclass
df1['call_name'] = t3.call_name
df1
```

```python
from sklearn.model_selection import train_test_split

#독립변수, 종속 변수 분리

y_df = df1['survived'] #종속 변수
X_df = df1.drop('survived',axis = 1) # 독립변수

# 학습용 데이터와 평가용 데이터를 8:2 혹은 7:3으로 분리

X_train, X_test, y_train, y_test = train_test_split(X_df,y_df,test_size =0.2, random_state= 11)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()


rf_model.fit(X_train,y_train)

# 예측값 추출
rf_pred = rf_model.predict(X_test)

# 예측값과 실제값 비교
accuracy_rf = accuracy_score(y_test,rf_pred).round(2)

lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
lr_pred = lr_model.predict(X_test)
accuracy_lr = accuracy_score(y_test,lr_pred).round(2)


print(f'rf 정확도 : {accuracy_rf},lr 정확도 : {accuracy_lr}')
```
