### Q. 10을 더하는 함수 add_10 과 두 객체의 합 add_two_obj 함수를 정의한 후 결과값이 각각 20과 20을 출력하세요.

```python
def add_10(x):
    return x+10

def add_two_obj(x,y):
    return x+y

print(add_10(10))
print(add_two_obj(10,10))
```

```python
def mul_10(x):
    return x*10
def div_two_obj(x,y):
    return round(x/y,4)

print(mul_10(20))
print(div_two_obj(100,10))
```

### Q. titanic 데이터셋을 seaborn에서 로딩한 후 블린 인덱싱을 이용하여 아래 사항들을 수행하세요.
- 나이가 10대(10~19세)인 승객만 따로 선택
- 나이가 10세 미만(0~9세)이고 여성인 승객만 따로 선택
- 나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택

```python
df_teen = df.loc[df['age']<20]
df_teen = df_teen.loc[df_teen['age']>=10]
print(df_teen)

df_girl = df.loc[df['age']<10]
df_girl = df_girl.loc[df_girl['sex']=='female']
print(df_girl)

df_young = df.loc[df['age']<10]
df_senior = df.loc[df['age']>=60]
df_final = pd.concat([df_young,df_senior],axis = 0,ignore_index=True)
df_final=df_final[['age','sex','alone']]
df_final
```

### Q. 최대값 - 최소값을 반환하는 사용자 함수를 min_max로 정의하고 age, fare 열로 구성되는 타이타닉 데이터프레임df 에 적용하여 결과와 자료형을 출력하세요. source : titanic = sns.load_dataset('titanic')

```python
def min_max(x):
    return x.max()-x.min()
    
    
titanic = sns.load_dataset('titanic')
df2 = titanic.loc[:,['age','sex','class','fare','survived']]
grouped = df2.groupby(['class'])

grouped.agg(lambda x :min_max(x))
```
