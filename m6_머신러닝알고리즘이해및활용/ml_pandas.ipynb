```python
def age_cat(x):
    cat = 0
    if x == np.nan:
        return np.nan
    elif x<10:
        cat = 1
    elif x<20:
        cat = 2
    elif x<30:
        cat = 3
    elif x<40:
        cat = 4
    elif x<50:
        cat = 5
    elif x<60:
        cat = 6
    else:
        cat = 7
    return cat
        
t_df['Age_cat'] = t_df.Age.apply(lambda x: age_cat(x))
t_df.head()
```
